```python
import numpy as np # for arrays
import pandas as pd # for dataframes
from sklearn.model_selection import train_test_split # used to split data to training and test data 
from sklearn.preprocessing import StandardScaler # for data pre-processing
from sklearn import svm # support vector machine model is used 
from sklearn.metrics import accuracy_score # for model evaluation
```

```python
parkinsons_data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/parkin pred/parkinsons.csv')
```

```python
X = parkinsons_data.drop(columns = ['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, random_state=2)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
X_train = scaler.transform(X_train) # range numbers in 1 to -1
X_test = scaler.transform(X_test)
```

```python
parkinson_model = svm.SVC(kernel = 'linear')
parkinson_model.fit(X_train, Y_train)
```

```python
X_train_pred = parkinson_model.predict(X_train)
train_acc = accuracy_score(Y_train, X_train_pred)
print("Accuracy on Training Set")
print(train_acc)
```

```python
X_test_pred = parkinson_model.predict(X_test)
test_acc = accuracy_score(X_test_pred,Y_test)
print("Accuracy on Test Set")
print(test_acc) 
```
